```python
clients_list = [
    {
        "id": 'ss',
        "name": 'Samsung'
    },
    {
        "id": 'nest',
        "name": 'Nestle'
    },
    {
        "id": 'abb',
        "name": 'Abbott'
    },
    {
        "id": 'mdl',
        "name": 'Mondelez'
    }
]

digitas_teams_list = [
    {
        "id": 'tt',
        "name": 'Technical Team',
    },
    {
        "id": 'pmt',
        "name": 'PM Team', 
    },
    {
        "id": 'dt',
        "name": 'Data Team'
    }
]

campaigns = [
    {
        "name": "camp_a",
        "client": 'ss',
        "teams": ['tt']
    },
    {
        "name": "camp_b",
        "client": 'abb',
        "teams": ['tt', 'pmt']
    },
    {
        "name": "camp_c",
        "client": 'nest',
        "teams": ['tt', 'dt']
    },
    {
        "name": "camp_d",
        "client": 'mdl',
        "teams": []
    },
    {
        "name": "camp_e",
        "client": 'ss',
        "teams": ['pmt']
    },
]
```

```python
#Code từ tìm hiểu
 #Lấy tên camp và tên team bằng ID
camps_name_1 = [camp["name"] for camp in campaigns if team_id in camp["teams"]]
team_name_1 = [team["name"] for team in digitas_teams_list if team_id in team["id"]]
client_name_1 = [client["name"] for client in clients_list if client_id == client["id"]]

#Input
team_id = ['tt',"pmt"]
client_id = 'mdl'

print(client_name_1)
print(team_name_1)
print(camps_name_1)
#Output
if client_name_1 and team_name_1:
    print(f'The associated campaigns of { ", ".join(client_name_1)} were handled by {", " .join(team_name_1)}: {", ".join(camps_name_1)}')
    print(f'The associated campaigns for {team_name_1[0]} are: {", ".join(camps_name_1)}')
else:
    print("No associated data found.")
print("-----------")

#code chatgpt

# Lấy tên camp và tên team bằng ID
camps_name_1 = [camp["name"] for camp in campaigns if any(t_id in camp["teams"] for t_id in team_id)]
team_name_1 = [team["name"] for team in digitas_teams_list if team["id"] in team_id]
client_name_1 = [client["name"] for client in clients_list if client_id == client["id"]]

# Input
team_id = ['tt', "pmt"]
client_id = 'mdl'

# In ra các tên đã lấy
print(client_name_1)
print(team_name_1)
print(camps_name_1)

# Output
if client_name_1 and team_name_1:
    print(f'The associated campaigns of { ", ".join(client_name_1)} were handled by {", " .join(team_name_1)}: {", ".join(camps_name_1)}')
    print(f'The associated campaigns for {team_name_1[0]} are: {", ".join(camps_name_1)}')
else:
    print("No associated data found.")
print("-----------------------")

# Code anh Đức chỉ

def get_camps_by_team_id(team_id):
    camps = []
    for camp in campaigns:
        if team_id in camp["teams"]:
            camps.append(camp["name"])
    return camps

#Lấy tên Team
def get_team_name_by_id(team_id):
    for team in digitas_teams_list:
        if team_id == team["id"]:
            return team["name"]

#Lấy tên Client
def get_client_name_by_id(client_id):
    for client in clients_list:
        if client["id"]==client_id:
            return client["name"]

#Input
team_id = 'tt'
client_id = 'ss'

#rename def
camps_name = get_camps_by_team_id(team_id)
team_name = get_team_name_by_id(team_id)
client_name = get_client_name_by_id(client_id)

#Output
print(f'The associated campaigns of {client_name} were handle by {team_name}: {", ".join(camps_name)}')

print(f'The associated campaigns for {team_name} are: {", ".join(camps_name)}')
```
