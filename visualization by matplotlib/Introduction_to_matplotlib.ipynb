### 2 ways to use plt: 1. Pyplot API vs Object_Oriented API
* Pyplot API ----> Quickly
* OO API ---> Advanced

# Pyplot API

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="pink");
plt.xlabel("Bien X")
plt.ylabel("Sin X")
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="pink");
plt.xlabel("Bien X")
plt.ylabel("Sin X")
plt.xlim([0,4])
plt.ylim([-.7,-.6])
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color="pink", linestyle="dashed", label="sin(x)");
plt.plot(x,np.cos(x), color="blue", label="cos(x)" );

plt.title("Sin(X) & Cos(x)")
plt.xlabel("Bien X");
plt.ylabel("F(x)");
plt.legend();
#plt.axis([0,4,-0,7, 0.6])
#plt.axis([xmin, xmax, ymin, ymax])
#plt.axis("equal")
#plt.axis("tight")
```

##OO API

```python
#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot

#3. Plot the data
ax.plot(x,y)

#4. Customize the plot
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis"
       );
```

### Line chart

```python
#create an array
x = np.linspace(0,10,100)
x[:5];
```

### Scatter Chart

```python
#Pyplot API
plt.scatter(x,np.exp(x)); #y=e^x
```

```python
#OO API

#Prepare New Data

rng = np.random.RandomState(0) # to hm random

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()

img1 = ax.scatter(x,y, s=sizes, c=colors, alpha=0.5
        )
fig.colorbar(img1)
```

### Bar Chart
* Vertical
* Horizontal

```python
#Prepare Data From python dictionary
soft_drink_prices = {"coke" : 10,
                     "Pessi": 12,
                     "Aqua": 8
                     }

fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="VinMart soft drink prices", ylabel="Price($)", xlabel="Sn phm")
```

```python
fig, ax = plt.subplots()
ax.barh(soft_drink_prices.keys(), soft_drink_prices.values());
```

### Histogram

```python
#Prepare data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:5]
```

```python
#Pyplot API

plt.hist(student_height, bins = 30);
```

# Subplots

* OO API: plt.subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2, 
                       ncols= 2, 
                       figsize = (10, 5))

#plot data to each axis

ax1.plot(x,x/2)
ax1.plot(x, x*2)

ax2.scatter(np.random.random(100), np.random.random(100))

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax4.hist(student_height, bins = 30);
```

# Option 2: 

```python
fig, ax = plt.subplots(nrows=2, 
                       ncols= 2, 
                       figsize = (10,5))

ax[0,0].plot(x,x/2)
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.random(100), np.random.random(100))

ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax[1,1].hist(student_height, bins = 30);
```
