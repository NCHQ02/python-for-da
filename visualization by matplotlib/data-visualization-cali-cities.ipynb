### Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read Dataset
raw_cali_cities =pd.read_csv("./california_cities.csv")
```

```python
raw_cali_cities.head()
```

```python
#Extract latd and Longd

cities = raw_cali_cities

lat, lon = cities["latd"], cities["longd"]

#Extract Population and Area
population, area = cities["population_total"], cities["area_total_km2"]


#plot using Pyplot API

#Initialize the Plot
plt.figure(figsize=(8,6))

#Scatter Plot Creation
plt.scatter(lon, lat, c=np.log10(population), s=area );

#Customize Axes
plt.axis("equal")
plt.xlabel("longatitude")
plt.ylabel("latitude");

#Colorbar for Population
plt.colorbar(label="log$_{10}$(population)");
plt.clim(3,7);

#Add title
plt.title("California Cities: Population and Area Distribution")

#Create a lengend for cities'size
area_range = [50, 100, 300, 500]

for area in area_range: 
    plt.scatter([],[], c="k", alpha=0.4, s=area, label=str(area) + "km$^2$")

plt.legend(scatterpoints=1, labelspacing=1,title="City Area");
```
