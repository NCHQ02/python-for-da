## Data Imputation (Missing Data Replacement)
Datasets often have missing values and this can cause problems for machine learning algorithms. It is considered good practise to identify and replace missing value in each column of your dataset prior to performing predictive modelling. This methond of missing data replacement is referred to as data inputation

```python
for col in data_df.columns:
    
    missing_data = data_df[col].isna().sum()
    
    missing_percent = missing_data/len(data_df)    
    print(f"Column {col}: has {missing_data} missing data")
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.heatmap(data_df.isna(), cmap="Blues");
```

```python
from sklearn.impute import SimpleImputer
```

```python
#Create an instance of Class SimpleImputer: np.nan is the empty value in the dataset
imputer = SimpleImputer(missing_values=np.nan,strategy="mean") 
```
